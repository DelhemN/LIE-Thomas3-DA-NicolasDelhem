```python
import os
import chainlit as cl
from langchain import HuggingFaceHub, PromptTemplate, LLMChain
from getpass import getpass

# Prompt the user to enter the Hugging Face Hub API token securely
huggingfacehub_api_token = getpass("Enter your Hugging Face Hub API token: ")

# Set the environment variable with the entered token
os.environ["HUGGINGFACEHUB_API_TOKEN"] = huggingfacehub_api_token

model_id = "sshleifer/tiny-python-gpt"
python_master_model = HuggingFaceHub(
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"],
    repo_id=model_id,
    model_kwargs={"temperature": 0.8, "max_new_tokens": 200}
)

template = """ 
You are PythonMaster, a knowledgeable AI assistant specializing in Python.

User Query: {python_query} 
"""

prompt = PromptTemplate(template=template, input_variables=['python_query'])

python_master_chatbot = LLMChain(llm=python_master_model, prompt=prompt, verbose=True)


#Initialize 

user_query = "How do I read a CSV file in Python?"
response = python_master_chatbot.run(python_query = user_query)

print(response)
```
